```python
import numpy as np
import tensorflow as tf
```

```python
iterator = tf.python_io.tf_record_iterator(path='../src/train-fold0-0.tfrecords')
```

```python
for idx, record in enumerate(iterator):
    example = tf.train.Example()
    example.ParseFromString(record)
    img_raw = example.features.feature['image_raw'].bytes_list.value[0]
    mask_raw = example.features.feature['mask_raw'].bytes_list.value[0]
    if idx == 10:
        break
    
```

```python
img = np.frombuffer(img_raw, dtype=np.float32).reshape(128, 128)
mask = np.frombuffer(mask_raw, dtype=np.uint8).reshape(128, 128)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()
```

```python
plt.imshow(mask)
plt.show()
```
